```python

# ── Preparar datos ────────────────────────────────────────────────────────────
df_año = (
    a24.groupby("AÑO ESTRENO")
    .size()
    .reset_index(name="total_peliculas")
    .sort_values("AÑO ESTRENO")
)

df_año["AÑO ESTRENO"] = df_año["AÑO ESTRENO"].astype(int)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = go.Figure()

fig.add_trace(
    go.Bar(
        x=df_año["AÑO ESTRENO"],
        y=df_año["total_peliculas"],
        marker=dict(
            color=df_año["total_peliculas"],
            colorscale="Blues",
            showscale=False,
        ),
        text=df_año["total_peliculas"],
        textposition="outside",
        textfont=dict(color="white", size=11),
        hovertemplate="<b>Año: %{x}</b><br>Películas: %{y}<extra></extra>",
    )
)

# ── Layout ────────────────────────────────────────────────────────────────────
fig.update_layout(
    title=dict(
        text="<b>PELÍCULAS POR AÑO DE ESTRENO</b><br><sup>Total de estrenos por año</sup>",
        font=dict(size=22, color="white", family="Arial Black"),
        x=0.5,
        xanchor="center",
        y=0.97,
    ),

    paper_bgcolor="#0f0f1a",
    plot_bgcolor="#0f0f1a",

    xaxis=dict(
        title="Año de Estreno",
        color="white",
        gridcolor="rgba(255,255,255,0.05)",
        tickmode="linear",       # muestra todos los años
        tickangle=-45,
        dtick=1,
    ),
    yaxis=dict(
        title="Total de Películas",
        color="white",
        gridcolor="rgba(255,255,255,0.1)",
    ),

    font=dict(color="white"),
    margin={"r": 40, "t": 100, "l": 60, "b": 80},
    height=550,

    annotations=[
        dict(
            text="Elaboración propia",
            x=0.99, y=0.01,
            xref="paper", yref="paper",
            showarrow=False,
            font=dict(size=10, color="rgba(255,255,255,0.4)"),
            xanchor="right",
        )
    ],
)

fig.show()
```

```python
import plotly.graph_objects as go

# ── Preparar datos ────────────────────────────────────────────────────────────
df_tiempo = (
    a24.groupby("AÑO ESTRENO")
    .agg(
        presupuesto=("PRESUPUESTO",     "sum"),
        taquilla   =("TAQUILLA MUNDIAL","sum"),
    )
    .reset_index()
    .sort_values("AÑO ESTRENO")
)

df_tiempo["AÑO ESTRENO"] = df_tiempo["AÑO ESTRENO"].astype(int)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = go.Figure()

# Línea — Presupuesto
fig.add_trace(
    go.Scatter(
        x=df_tiempo["AÑO ESTRENO"],
        y=df_tiempo["presupuesto"],
        name="Presupuesto",
        mode="lines+markers",
        line=dict(color="cornflowerblue", width=4),
        marker=dict(size=8, color="cornflowerblue"),
        hovertemplate="<b>%{x}</b><br>Presupuesto: $%{y:,.0f}<extra></extra>",
    )
)

# Línea — Taquilla Mundial
fig.add_trace(
    go.Scatter(
        x=df_tiempo["AÑO ESTRENO"],
        y=df_tiempo["taquilla"],
        name="Taquilla Mundial",
        mode="lines+markers",
        line=dict(color="limegreen", width=4),
        marker=dict(size=8, color="limegreen"),
        hovertemplate="<b>%{x}</b><br>Taquilla: $%{y:,.0f}<extra></extra>",
    )
)

# ── Layout ────────────────────────────────────────────────────────────────────
fig.update_layout(
    title=dict(
        text="<b>PRESUPUESTO VS TAQUILLA MUNDIAL POR AÑO</b><br><sup>Escala logarítmica — Comparativa de inversión y recaudación</sup>",
        font=dict(size=22, color="white", family="Arial Black"),
        x=0.5,
        xanchor="center",
        y=0.97,
    ),

    paper_bgcolor="#0f0f1a",
    plot_bgcolor="#0f0f1a",

    xaxis=dict(
        title="Año de Estreno",
        color="white",
        gridcolor="rgba(255,255,255,0.05)",
        tickmode="linear",
        tickangle=-45,
        dtick=1,
    ),
    yaxis=dict(
        title="Monto (USD) — Escala Logarítmica",
        color="white",
        gridcolor="rgba(255,255,255,0.1)",
        type="log",                          # ✅ escala logarítmica
        tickformat="$,.0f",
        exponentformat="none",               # muestra números completos, no notación científica
    ),

    legend=dict(
        font=dict(color="white", size=12),
        bgcolor="rgba(15,15,35,0.85)",
        bordercolor="rgba(255,255,255,0.2)",
        borderwidth=1,
        x=0.01,
        y=0.99,
        xanchor="left",
        yanchor="top",
    ),

    font=dict(color="white"),
    margin={"r": 40, "t": 100, "l": 80, "b": 80},
    height=580,

    annotations=[
        dict(
            text="Elaboración propia",
            x=0.99, y=0.01,
            xref="paper", yref="paper",
            showarrow=False,
            font=dict(size=10, color="rgba(255,255,255,0.4)"),
            xanchor="right",
        )
    ],
)

fig.show()
```

```python
import plotly.graph_objects as go

# ── Preparar datos ────────────────────────────────────────────────────────────
df_dir = (
    a24.groupby("DIRECTOR")
    .agg(
        taquilla   =("TAQUILLA MUNDIAL", "sum"),
        presupuesto=("PRESUPUESTO",      "sum"),
    )
    .reset_index()
)

# Top 10 por taquilla
top_taquilla = (
    df_dir.sort_values("taquilla", ascending=False)
    .head(10)
    .sort_values("taquilla", ascending=True)
)

# Top 10 por presupuesto
top_presupuesto = (
    df_dir.sort_values("presupuesto", ascending=False)
    .head(10)
    .sort_values("presupuesto", ascending=True)
)

# ── Figura ────────────────────────────────────────────────────────────────────
fig = go.Figure()

# Traza 1 — Taquilla (visible por defecto)
fig.add_trace(
    go.Bar(
        x=top_taquilla["taquilla"],
        y=top_taquilla["DIRECTOR"],
        orientation="h",
        name="Taquilla Mundial",
        visible=True,
        marker=dict(
            color=top_taquilla["taquilla"],
            colorscale="Blues",
            showscale=False,
        ),
        text=top_taquilla["taquilla"].apply(lambda x: f"${x:,.0f}"),
        textposition="outside",
        textfont=dict(color="white", size=11),
        hovertemplate="<b>%{y}</b><br>Taquilla: $%{x:,.0f}<extra></extra>",
    )
)

# Traza 2 — Presupuesto (oculta por defecto)
fig.add_trace(
    go.Bar(
        x=top_presupuesto["presupuesto"],
        y=top_presupuesto["DIRECTOR"],
        orientation="h",
        name="Presupuesto",
        visible=False,
        marker=dict(
            color=top_presupuesto["presupuesto"],
            colorscale="Greens",
            showscale=False,
        ),
        text=top_presupuesto["presupuesto"].apply(lambda x: f"${x:,.0f}"),
        textposition="outside",
        textfont=dict(color="white", size=11),
        hovertemplate="<b>%{y}</b><br>Presupuesto: $%{x:,.0f}<extra></extra>",
    )
)

# ── Botones ───────────────────────────────────────────────────────────────────
botones = [
    dict(
        label="🎬 Taquilla Mundial",
        method="update",
        args=[
            {"visible": [True, False]},
            {"title": {
                "text": (
                    "<b>TOP 10 DIRECTORES POR TAQUILLA MUNDIAL</b><br>"
                    "<sup>Total de recaudación acumulada</sup>"
                )
            },
            "xaxis": {"title": "Taquilla Mundial (USD)", "tickformat": "$,.0f",
                      "color": "white", "gridcolor": "rgba(255,255,255,0.1)"}
            },
        ],
    ),
    dict(
        label="💰 Presupuesto",
        method="update",
        args=[
            {"visible": [False, True]},
            {"title": {
                "text": (
                    "<b>TOP 10 DIRECTORES POR PRESUPUESTO</b><br>"
                    "<sup>Total de inversión acumulada</sup>"
                )
            },
            "xaxis": {"title": "Presupuesto (USD)", "tickformat": "$,.0f",
                      "color": "white", "gridcolor": "rgba(255,255,255,0.1)"}
            },
        ],
    ),
]

# ── Layout ────────────────────────────────────────────────────────────────────
fig.update_layout(
    updatemenus=[
        dict(
            buttons=botones,
            direction="right",
            showactive=True,
            x=0.5,
            xanchor="center",
            y=1.12,
            yanchor="top",
            bgcolor="#1a1a2e",
            bordercolor="rgba(255,255,255,0.3)",
            font=dict(color="white", size=12),
        )
    ],

    title=dict(
        text=(
            "<b>TOP 10 DIRECTORES POR TAQUILLA MUNDIAL</b><br>"
            "<sup>Total de recaudación acumulada</sup>"
        ),
        font=dict(size=22, color="white", family="Arial Black"),
        x=0.5,
        xanchor="center",
        y=0.97,
    ),

    paper_bgcolor="#0f0f1a",
    plot_bgcolor="#0f0f1a",

    xaxis=dict(
        title="Taquilla Mundial (USD)",
        color="white",
        gridcolor="rgba(255,255,255,0.1)",
        tickformat="$,.0f",
    ),
    yaxis=dict(
        color="white",
        automargin=True,
    ),

    font=dict(color="white"),
    margin={"r": 180, "t": 120, "l": 20, "b": 60},
    height=500,

    annotations=[
        dict(
            text="Elaboración propia",
            x=0.99, y=-0.02,
            xref="paper", yref="paper",
            showarrow=False,
            font=dict(size=10, color="rgba(255,255,255,0.4)"),
            xanchor="right",
        )
    ],
)

fig.show()
```

```python
import plotly.graph_objects as go

# ── Preparar datos — ROI por director ────────────────────────────────────────
df_roi = (
    a24.groupby("DIRECTOR")
    .agg(
        taquilla   =("TAQUILLA MUNDIAL", "sum"),
        presupuesto=("PRESUPUESTO",      "sum"),
    )
    .reset_index()
)

df_roi["ROI"] = df_roi["taquilla"] / df_roi["presupuesto"]

# Top 10 por ROI
df_roi = (
    df_roi.sort_values("ROI", ascending=False)
    .head(10)
    .sort_values("ROI", ascending=True)
)

# ── Color según ROI (verde si > 1, rojo si < 1) ───────────────────────────────
colores = ["limegreen" if r >= 1 else "tomato" for r in df_roi["ROI"]]

# ── Figura ────────────────────────────────────────────────────────────────────
fig = go.Figure(
    go.Bar(
        x=df_roi["ROI"],
        y=df_roi["DIRECTOR"],
        orientation="h",
        marker=dict(
            color=colores,
            line=dict(width=1, color="rgba(255,255,255,0.1)"),
        ),
        text=df_roi["ROI"].apply(lambda x: f"{x:.2f}x"),
        textposition="outside",
        textfont=dict(color="white", size=12),
        hovertemplate=(
            "<b>%{y}</b><br>"
            "ROI: %{x:.2f}x<br>"
            "<extra></extra>"
        ),
    )
)

# Línea de referencia ROI = 1 (punto de equilibrio)
fig.add_vline(
    x=1,
    line=dict(color="rgba(255,255,255,0.5)", width=2, dash="dash"),
    annotation_text="Punto de equilibrio",
    annotation_font=dict(color="rgba(255,255,255,0.6)", size=11),
    annotation_position="top",
)

# ── Layout ────────────────────────────────────────────────────────────────────
fig.update_layout(
    title=dict(
        text="<b>TOP 10 DIRECTORES POR ROI</b><br><sup>Taquilla Mundial ÷ Presupuesto — Mayor retorno de inversión</sup>",
        font=dict(size=22, color="white", family="Arial Black"),
        x=0.5,
        xanchor="center",
        y=0.97,
    ),

    paper_bgcolor="#0f0f1a",
    plot_bgcolor="#0f0f1a",

    xaxis=dict(
        title="ROI (Taquilla / Presupuesto)",
        color="white",
        gridcolor="rgba(255,255,255,0.1)",
        tickformat=".1f",
        ticksuffix="x",
    ),
    yaxis=dict(
        color="white",
        automargin=True,
    ),

    font=dict(color="white"),
    margin={"r": 100, "t": 100, "l": 20, "b": 60},
    height=500,

    annotations=[
        dict(
            text="Elaboración propia",
            x=0.99, y=-0.02,
            xref="paper", yref="paper",
            showarrow=False,
            font=dict(size=10, color="rgba(255,255,255,0.4)"),
            xanchor="right",
        )
    ],
)

fig.show()
```
